```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_excel("Food_Establishment_Inspection_Scores_(1).xlsx")

data.head()
```

```python
data["Street"] = [address.split('\n')[0] for address in data["Address"]]
data["City"] = [address.split('\n')[1].split(",")[0] for address in data["Address"]]
data["State"] = [address.split('\n')[1].split(",")[1].strip().split(" ")[0] for address in data["Address"]]
data["ZipCode"] = [address.split('\n')[1].split(",")[1].strip().split(" ")[1] for address in data["Address"]]

geo = []
for address in data["Address"]:
    if len(address.split('\n')) == 3:
        geo.append(address.split('\n')[2])
    else:
        geo.append("")
data["GEO"] = geo


#for address in data["Address"]:
    #if address.split('\n')[1] 
#    print(address)
#    if "," not in address.split('\n')[1]:
#        print(address.split('\n')[1].split(",")[1].strip().split(" ")[0])
```
