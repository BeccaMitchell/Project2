```python
import pandas as pd
import matplotlib.pyplot as plt
data = pd.read_excel("Food_Establishment_Inspection_Scores_(2).xlsx")

data.head()
```

```python
business = data[['Restaurant Name', 'Zip Code', 'Address', 'Facility ID', 'Street', 'City', 'State', 'ZipCode', 'GEO']]
print(len(business))
business = business.drop_duplicates(subset='Facility ID')
print(len(business))
business.head()
```

```python
business.to_excel("Food_Establishment_Unique.xlsx", index=False)
```

```python
business_scores = data[['Inspection Date', 'Score', 'Facility ID']]
business_scores = business_scores.sort_values(by=['Facility ID', 'Inspection Date'], ascending=[1,0])
business_scores.head()
```

```python
business_scores.to_excel("Food_Establishment_Inspections.xlsx", index=False)
```
